## Extracción

```python
df = pd.DataFrame(columns = ['id', 'emotion', 'intensity', 'confidence', 'lead_time', 'time_interval', 'channel'])
df.head()
```

```python
for i in range(len(data)):
    identity = data[i]['id']
    emotion = data[i]['annotations'][0]['result'][0]['value']['choices'][0]
    intensity = data[i]['annotations'][0]['result'][1]['value']['choices'][0]
    confidence = data[i]['annotations'][0]['result'][2]['value']['choices'][0]
    lead_time = data[i]['annotations'][0]['lead_time']
    channel = int(data[i]['data']['url'][-10])
    
    # tiempo entre create y update, en segundos
    created_str = data[i]['annotations'][0]['created_at'].replace('T', ' ')[:-8]
    updated_str = data[i]['annotations'][0]['updated_at'].replace('T', ' ')[:-8]
    created_time = datetime.datetime.strptime(created_str, format)
    updated_time = datetime.datetime.strptime(updated_str, format)
    time_interval = (updated_time - created_time).total_seconds()
    
    fila = [identity, emotion, intensity, confidence, lead_time, time_interval, channel]
    
    df.loc[len(df)] = fila
    
df.head()
```
